```python
#driver 실행

from selenium import webdriver
import time

chrome_path = "c://tmp//chromedriver.exe"

options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(chrome_path,options=options)
time.sleep(3)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import re

driver.get("https://search.naver.com/search.naver?where=news&query=%EC%A3%BC%EC%B0%A8%EA%B5%AC%EC%97%AD&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds=2021.12.01&de=2022.02.28&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20211201to20220228&is_sug_officeid=0")
time.sleep(2)

date_list = []
title_list = []
content_list = []

while True:
    try:
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        #제목
        titles = soup.find('div','group_news').find_all('a','news_tit')
        for title in titles:
            title = re.sub('[^가-힣\s]','',title.text)
            title_list.append(title)

        #내용
        contents = soup.find('div','group_news').find_all('a','api_txt_lines dsc_txt_wrap')
        for content in contents:
            content = re.sub('[^가-힣\s]','',content.text)
            content_list.append(content)
    
        #날짜
        dates = soup.find_all('div','info_group')
        for date in dates:
            date_list.append(date.text)

        driver.find_element_by_class_name("btn_next").click()
        time.sleep(2)
    except:
        break
```

```python
import pandas as pd

df = pd.DataFrame(data={'title':title_list,'content':content_list,'date':date_list})
len(df)
```
